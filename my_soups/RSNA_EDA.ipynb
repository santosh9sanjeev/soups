```python
import random
random.seed(42)  # for reproducibility
merged_df_v1['random_number'] = [random.random() for _ in range(len(merged_df_v1))]

# Step 2: Assign 'train', 'val', or 'test' based on probabilities
merged_df_v1['split'] = pd.cut(merged_df_v1['random_number'], bins=[0, 0.8, 0.9, 1], labels=['train', 'val', 'test'])
merged_df_v1.drop(columns=['random_number'], inplace=True)

# Step 3: Calculate counts
counts = merged_df_v1.groupby(['split', 'class']).size().unstack(fill_value=0)
total_counts = counts.sum(axis=1)

print("Counts by split and target:")
print(counts)

print("\nTotal counts by split:")
print(total_counts)
```
