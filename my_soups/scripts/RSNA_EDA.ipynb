```python
df = pd.read_csv('stage_2_train_labels.csv')

# Find duplicate patientIDs
duplicate_patientIDs = df[df.duplicated(subset=['patientId'], keep=False)]

# Create a new DataFrame with rows containing duplicate patientIDs
duplicates_df = df[df['patientId'].isin(duplicate_patientIDs['patientId'])]

# Remove the duplicates from the original DataFrame
df = df[~df['patientId'].isin(duplicate_patientIDs['patientId'])]
```

```python
import random
random.seed(42)  # for reproducibility
df_temp1['random_number'] = [random.random() for _ in range(len(df_temp1))]

# Step 2: Assign 'train', 'val', or 'test' based on probabilities
df_temp1['split'] = pd.cut(df_temp1['random_number'], bins=[0, 0.8, 0.9, 1], labels=['train', 'val', 'test'])
df_temp1.drop(columns=['random_number'], inplace=True)

# Step 3: Calculate counts
counts = df_temp1.groupby(['split', 'Target']).size().unstack(fill_value=0)
total_counts = counts.sum(axis=1)

print("Counts by split and target:")
print(counts)

print("\nTotal counts by split:")
print(total_counts)
```
